## Focus
I would like to see the relation between gender, family status and total ATM incom.

```python
subdata = app_train.loc[:,('CODE_GENDER', 'AMT_INCOME_TOTAL', 'NAME_FAMILY_STATUS')]
subdata
```

Get the mean of total ATM incom by the group (gender, family status).

```python
# status = subdata.iloc[:, 2].unique()
# print(status)
summary_df = subdata.groupby(['CODE_GENDER','NAME_FAMILY_STATUS']).mean()
display(summary_df) # pretty dataframe, not use 'print'
```

```python
status = summary_df.index.get_level_values(1)[0:6].values.tolist()
status
```

```python
# set new index put missing value with zero
new_index = pd.MultiIndex.from_product([ ['F', 'M', 'XNA'], status ],
                                      names=['gender', 'status'])
summary_df = summary_df.reindex(new_index, fill_value = 0)
summary_df.rename(index={'Single / not married': 'Single'}, inplace = True)
status = summary_df.index.get_level_values(1)[0:6].values.tolist()
display(summary_df)
```

```python
# get the means for each gender
F_mean = summary_df.loc['F',:].values.flatten().tolist()
M_mean = summary_df.loc['M',:].values.flatten().tolist()
XNA_mean = summary_df.loc['XNA',:].values.flatten().tolist()
```

## Plot the Result

```python
x = np.arange(len(status))  # the label locations
width = 0.3  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, F_mean, width, label = 'Women')
rects2 = ax.bar(x        , M_mean, width, label = 'Men')
rects3 = ax.bar(x + width, XNA_mean, width, label = 'NA')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Total ATM Income / ten thousand')
ax.set_title('Total ATM Income by family status and gender')
ax.set_xticks(x)
ax.set_xticklabels(status)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(round(height/10000, 1)),
                    xy=(rect.get_x() + rect.get_width()/2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
        
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.show()
```

We can see that male with unknown family status get the most total ATM income in this data. And male's mean value is higher than female in all of the six cases.
